```
# código à respeito do gráfico "Em qual rede de ensino você estuda?"

# Número Randômico para a aleatoriedade
np.random.seed(19680801)

plt.rcdefaults()
fig, ax = plt.subplots()

# Opções do ensino Público e Privado
ensino = ('Pública', 'Privada')
y_pos = np.arange(len(ensino))
quantidade_pessoas = [12,2] 
error = np.random.rand(len(ensino))

ax.barh(y_pos, quantidade_pessoas,height=0.5, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(ensino)
ax.invert_yaxis()  
ax.set_xlabel('Alunos')
ax.set_title('Quantidade de alunos por Ensino Público e Privado')

plt.xticks([0.0, 2.5, 5.0, 7.5, 10.0, 12.5]) 

plt.show()
```

```
# código à respeito do gráfico "Você tem conseguido estudar na quarentena?"
import matplotlib.pyplot as plt #importando as bibliotecas Matplotlib e Numpy para construir os gráficos
import numpy as np


# Número Randômico para a aleatoriedade
np.random.seed(19680801)

plt.rcdefaults()
fig, ax = plt.subplots()

# Opções do tipo de Ensino
ensino = ('Sim', 'Não', 'Às vezes')
y_pos = np.arange(len(ensino))
quantidade_pessoas = [9, 1, 3] 
error = np.random.rand(len(ensino))

ax.barh(y_pos, quantidade_pessoas,height=0.5, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(ensino)
ax.invert_yaxis()  
ax.set_xlabel('Quantidade de Alunos')
ax.set_title('Êxito de Estudo na Quarentena')

plt.xticks([0, 2, 4, 6, 8, 10]) #modifica o eixo x

plt.show()
```

```
# código à respeito do gráfico "Qual tipo de ambiente para estudo você tem em casa?"
import matplotlib.pyplot as plt #importando as bibliotecas Matplotlib e Numpy para construir os gráficos
import numpy as np


# Número Randômico para a aleatoriedade
np.random.seed(19680801)

plt.rcdefaults()
fig, ax = plt.subplots()

# Opções do tipo de Ensino
ensino = ('Ambiente privativo, confortável e silencioso', 'Ambiente compartilhado e/ou não silencioso', 'Não possui ambiente específico para estudo')
y_pos = np.arange(len(ensino))
quantidade_pessoas = [7, 5, 1] 
error = np.random.rand(len(ensino))

ax.barh(y_pos, quantidade_pessoas,height=0.5, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(ensino)
ax.invert_yaxis()  
ax.set_xlabel('Quantidade de Alunos')
ax.set_title('Tipo de Ambiente de Estudos dos Alunos')

plt.xticks([0, 2, 4, 6, 8]) #modifica o eixo x

plt.show()
```

```
# código à respeito do gráfico "Como você considera a qualidade dos seus estudos??"
import matplotlib.pyplot as plt

data = {'Muito Ruim': 1, '2': 1, '3': 3, '4': 4, 'Muito bom': 5}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots()
axs.bar(names, values)
fig.suptitle('Qualidade dos Estudos')
```
