```
#  MESMO CÓDIGO, MAS GRÁFICOS DIFERENTES, ESTAVA TESTANDO #Gente, acho que esse ficou estranho, não acham?
# código à respeito do gráfico "Que tipo de internet você tem disponível em sua residência?""
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["12 Fixa por assinatura (fibra, cabo ou rádio, própria ou compartilhada) ",
          "0 Disponível no meu smartphone quando tenho crédito (3G/4G)",
          "1 Sempre disponível no meu smartphone (3G/4G)",
          "0 Disponível no smartphone de outros quando tem crédito (3G/4G)"
          "0 Sempre disponível no smartphone de outros (3G/4G)"
          "0 Não possuo acesso à internet"
          "0 Outros"]

data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Tipos de Internet")

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Tipos de internet disponíveis nas residências dos alunos")

plt.show()
```

```
#  MESMO CÓDIGO, MAS GRÁFICOS DIFERENTES, ESTAVA TESTANDO
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["Fixa por assinatura (fibra, cabo ou rádio, própria ou compartilhada) ",
          "Disponível no meu smartphone quando tenho crédito (3G/4G)",
          "Sempre disponível no meu smartphone (3G/4G)",
          "Disponível no smartphone de outros quando tem crédito (3G/4G)",
          "Sempre disponível no smartphone de outros (3G/4G)",
          "Não possuo acesso à internet",
          "Outros"]

data = [12, 0, 1, 0, 0, 0, 0]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(7*np.sign(x), 7*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Tipos de internet disponíveis nas residências dos alunos")

plt.show()
```
